# **Data Cleaning**

```python
df_copy = df.copy()
```

## *Impossible values*

```python
df_copy.loc[df_copy['exam_score']<0,'exam_score'] = np.nan
df_copy.loc[df_copy['exam_score']>100,'exam_score'] = np.nan
df_copy.loc[df_copy['attendance_percent']>100,'attendance_percent'] = np.nan
```

## *Fill NaN*

```python
print("NaN values Before:- \n")
print(df_copy.isna().sum())

print("\nNaN values after:-\n")
df_copy['study_hours_per_day'] = df_copy['study_hours_per_day'].fillna(df_copy['study_hours_per_day'].median())
df_copy['attendance_percent'] = df_copy['attendance_percent'].fillna(df_copy['attendance_percent'].median())
df_copy['sleep_hours'] = df_copy['sleep_hours'].fillna(df_copy['sleep_hours'].median())
df_copy['exam_score'] = df_copy['exam_score'].fillna(df_copy['exam_score'].median())
print(df_copy.isna().sum())
```

## *Drop Duplicates*

```python
print(f"Number of Duplicates in Data is {df_copy.duplicated().sum()}")
df_copy.drop_duplicates(inplace=True)
print(f"\nThe number of Duplicates after Dropping is {df_copy.duplicated().sum()}")
```

# **Data Preperation**

```python
X = df_copy[['study_hours_per_day','attendance_percent','sleep_hours','previous_score','has_tuition','assignments_completed']]
y = df_copy['exam_score']

print("The Shape of X is ",X.shape)
print("The Shape of y is ",y.shape)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.6,random_state=42)

print(f'The Size of training data is {X_train.shape} and {y_train.shape}')
print(f'\nThe Size of Testing data is {X_test.shape} and {y_test.shape}\n')
print(f"The number of NaN Values in training data is:- \n{X_train.isna().sum() + y_train.isna().sum()}\n")
print(f"The number of NaN Values in testing data is:- \n{X_test.isna().sum() + y_test.isna().sum()}")
```

# **Model**

## *Train*

```python
model = LinearRegression()
model.fit(X_train,y_train)
print("Model Trained")
```

## *Training Evaluation*

```python
print(f"Intercept (β₀): {model.intercept_:.4f}\n")
print("Feature Coefficients:")
for feature, coef in zip(X.columns, model.coef_):
    print(f"  {feature:22} → {coef:.4f}")
```

## *Test*

```python
pred = model.predict(X_test)

comparison = pd.DataFrame({
    "Actual Values" : y_test.values,
    'Pridicted Values' : pred.round(3),
    'Difference': (y_test.values - pred).round(3)
})

print(comparison.head(10))
```

### *Accuracy*

```python
r2 = r2_score(y_test,pred)
print(f"The Accuracy of the model is {round(r2,4)*100}%")
```

### *Test Mectrics*

```python
mae  = mean_absolute_error(y_test, pred)
rmse = np.sqrt(mean_squared_error(y_test, pred))

print(f"On an Average the Model is Off by:  {mae:.3f} Marks")
print(f"RMSE:      {rmse:.4f}")
```
